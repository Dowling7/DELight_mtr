```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft, rfftfreq
from scipy.interpolate import interp1d
import os

# === Parameters ===
sampling_rate = 3906250
uv_template_path = "../templates/template_K_alpha_tight.npy"
qp_template_path = "../templates/template_qp_toy.npy"
mmc_path = "../templates/noise_psd_from_MMC.npy"
color_noise_dir = "../templates/color"
noise_types = ['white', 'pink', 'brownian', 'blue', 'violet']

# === Load UV and QP templates ===
if not os.path.exists(uv_template_path):
    raise FileNotFoundError(f"UV template not found at: {uv_template_path}")
if not os.path.exists(qp_template_path):
    raise FileNotFoundError(f"QP template not found at: {qp_template_path}")

template_uv = np.load(uv_template_path)
template_qp = np.load(qp_template_path)

# === Prepare FFT axis ===
N = len(template_uv)  # Assume both templates are the same length
frequencies = rfftfreq(N, d=1.0 / sampling_rate)

# === Compute Amplitude Spectral Density (FFT) ===
fft_uv = np.abs(rfft(template_uv)) / np.sqrt(sampling_rate * N)
fft_qp = np.abs(rfft(template_qp)) / np.sqrt(sampling_rate * N)

# === Load MMC PSD ===
if not os.path.exists(mmc_path):
    raise FileNotFoundError(f"MMC PSD not found at: {mmc_path}")
mmc_psd = np.load(mmc_path)

# Interpolate MMC PSD if needed
if len(mmc_psd) != len(frequencies):
    original_freq = np.linspace(frequencies[0], frequencies[-1], len(mmc_psd))
    mmc_psd = interp1d(original_freq, mmc_psd, bounds_error=False, fill_value="extrapolate")(frequencies)

# === Load theoretical colored noise PSDs ===
psds = {
    "UV Template FFT": fft_uv,
    "QP Template FFT": fft_qp,
    "MMC Noise PSD": mmc_psd
}

for noise in noise_types:
    file_path = os.path.join(color_noise_dir, f"noise_psd_{noise}.npy")
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Missing PSD file for {noise}: {file_path}")
    
    color_psd = np.load(file_path)
    psds[f"{noise.capitalize()} Noise PSD"] = color_psd

# === Plot all ===
plt.figure(figsize=(10, 6))

for label, psd_vals in psds.items():
    plt.loglog(frequencies, psd_vals, label=label)

plt.title("Noise PSD compare to FFT of signals")
plt.xlabel("Frequency [Hz]")
plt.ylabel("PSD (A²/Hz) and FFT (A/√Hz)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
