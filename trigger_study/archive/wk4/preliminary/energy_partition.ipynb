```python
import numpy as np
import matplotlib.pyplot as plt
from TraceSimulator.DELightSignalFormation import DELightSignalFormation

# === Step 1: Initialize the signal partition model ===
partition_path = '/ceph/bmaier/delight/share/TraceSimulator/signal_partition/'
dsf = DELightSignalFormation(template_path=partition_path, load='both')  # or 'ER', 'NR'

# === Step 2: Sample input energies (1 keV to 1000 keV) ===
energies = np.logspace(0, 3, 200)  # 1 keV to 1000 keV = 1 MeV
E_ph, E_tr, E_uv, E_ir = dsf.get_partition(energies, int_type='NR')  # 'ER' or 'NR'

# === Step 3: Calculate percentage partition ===
P_ph = (E_ph / energies) * 100
P_tr = (E_tr / energies) * 100
P_uv = (E_uv / energies) * 100
P_ir = (E_ir / energies) * 100

# === Step 4: Plot with log-scaled x-axis in keV ===
plt.figure(figsize=(10, 6))
plt.plot(energies, P_ph, label='Phonons')
plt.plot(energies, P_tr, label='Triplets')
plt.plot(energies, P_uv, label='UV Photons')
plt.plot(energies, P_ir, label='IR Photons')
plt.xscale('log')
plt.xlim(1, 1000)  # in keV
plt.title("Percentage Energy Partitioning for NR Events")
plt.xlabel("Deposited Energy [keV]")
plt.ylabel("Partitioned Energy [%]")
plt.ylim(0, 100)
plt.legend()
plt.grid(True, which='both', ls='--')
plt.tight_layout()
plt.show()
```
