```python
import numpy as np
import matplotlib.pyplot as plt
import helix as hx# Assuming hx is a library used in your context
template_path_1 = './plugins/event_rqs/default_templates.npy'
template_path_2 = '2level2_template.npy'

template_1 = np.load(template_path_1, allow_pickle=True)[1]
template_2 = np.load(template_path_2, allow_pickle=True)[0]

# Print dimensions and statistics
print("Template 1 shape:", template_1.shape)
print("Template 1 - Max:", np.max(template_1), "Min:", np.min(template_1), "Mean:", np.mean(template_1))
print("Template 2 shape:", template_2.shape)
print("Template 2 - Max:", np.max(template_2), "Min:", np.min(template_2), "Mean:", np.mean(template_2))
```

```python
# Plot templates
plt.figure(figsize=(12, 6))
plt.plot(template_1, label='Template 1', color='C1')
plt.plot(template_2, label='Template 2', color='C2', linestyle='dashed')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.title('Templates Comparison')
plt.legend()
plt.show()
```
