```python
import numpy as np

def duplicate_npy_template(input_file: str, output_file: str):
    """
    Reads a .npy file containing a (1, L) array and duplicates it to (2, L),
    then saves the new array to a new .npy file.

    :param input_file: Path to the input .npy file.
    :param output_file: Path to save the new .npy file.
    """
    # Load the original template
    array = np.load(input_file)
    
    if array.ndim == 1:
        array = array.reshape(1, -1)
    elif array.shape[0] != 1:
        raise ValueError("Input array must have shape (L,) or (1, L)")
    
    # Duplicate along axis 0 to create a (2, L) array
    duplicated_array = np.vstack([array, array])
    
    # Save the new template
    np.save(output_file, duplicated_array)
    print(f"New template saved to {output_file} with shape {duplicated_array.shape}")
```

```python
duplicate_npy_template('level2_template.npy', 'double_level2_template.npy')
```
